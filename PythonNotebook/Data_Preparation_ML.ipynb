```python
#Visualization before and after outlier removed
import matplotlib.pyplot as plt
```

```python
# Calculate Q1 and Q3
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)

# Calculate IQR
IQR = Q3 - Q1

# Define bounds for outliers
Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers *after* scaling
df_filtered= df[(df['Income'] >= Lower_bound) & (df['Income'] <= Upper_bound)]
```

```python
column_name='Income'
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))
df[column_name].hist(ax=ax[0],bins=30,color='skyblue',edgecolor='black')
ax[0].set_title('Before Outlier Remolval')
ax[0].set_xlabel('column_name')
ax[0].set_ylabel('Frequency')
df_filtered[column_name].hist(ax=ax[1],bins=30,color='lightgreen',edgecolor='black')
ax[1].set_title('After Outlier Removal')
ax[1].set_xlabel(column_name)
ax[1].set_ylabel('Frequency')
plt.show()
```
